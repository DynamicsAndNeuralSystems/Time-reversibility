```python
models = ['AR1_UNO', 'ARMA11_UNO', #ARMA
          'HEN', 'LOGISTIC4', 'QUADRATIC', # chaos
          'HEN_SUM', # sum chaos
          'MODA', 'LLOG', # other deterministic
          'SINE_STOCH'] # other stochastic

model_keywords = {'AR1_UNO': ['irreversible'], 'ARMA11_UNO': ['irreversible'],
                  'HEN': ['irreversible'], 'LOGISTIC4': ['irreversible'], 'QUADRATIC': ['irreversible'],
                  'HEN_SUM': ['irreversible'],
                  'MODA': ['irreversible'], 'LLOG': ['irreversible'],
                  'SINE_STOCH': ['irreversible']}
```

```python
num_ts=100
# Load the csv files
length = [10, 20, 50, 100, 200, 500, 1000, 2000,5000]
```

```python
# folder where this script lives
BASE_DIR = Path.cwd()
path_base = str(BASE_DIR)

REPO_DIR = BASE_DIR.parents[2]

DATA_DIR = REPO_DIR / 'data-tr'
print(DATA_DIR)
path_data= str(DATA_DIR)


# Directory to find hctsa pre-processed hctsa data
dir_hctsa= path_data + '/robustness-analysis/data-analysis-robustness/'
```

```python
# Identify common features across all the lengths (for some length some ops have been filtered out)

common_ops = set()
for i in length:
    df = pd.read_csv(dir_hctsa+f'/df_TS_DataMat_diff_{i}.csv')
    ops = set(df.columns) - {'Model'}
    if not common_ops:
        common_ops = ops
    else:
        common_ops.intersection_update(ops)

# Put in dataframe
df_common_ops = pd.DataFrame(columns=['Common ops'])
df_common_ops['Common ops'] = list(common_ops)
```

```python
# Load the hctsa csv files

dict_hctsa = {}
for i in length:
    df = pd.read_csv(dir_hctsa+f'/df_TS_DataMat_diff_{i}.csv')
    # Number in abs value (df except column Model)
    df.set_index('Model', inplace=True)
    #df = df.abs()
    # Select columns with features in common_ops
    df = df.loc[:, df.columns.intersection(common_ops)]
    dict_hctsa[i] = df
```

```python
# Print models
print("Models:", dict_hctsa[10].index.unique().tolist())
```

```python
# Select feature
op = 'AC_nl_001'

# Select model
model = 'SINE_STOCH' # other stochastic

avg_sine = []
std_sine = []
for i in length:
    df_hctsa = dict_hctsa[i]
    df_hctsa = df_hctsa.loc[model, op]
    # Mean
    avg = df_hctsa.mean()
    std = df_hctsa.std(ddof=1)
    avg_sine.append(avg)
    std_sine.append(std)
df_avg_sine = pd.DataFrame(columns=['Length', 'Mean'], data={'Length': length, 'Mean': avg_sine})
df_std_sine = pd.DataFrame(columns=['Length', 'Std'], data={'Length': length, 'Std': std_sine})

model = 'HEN' # chaos

avg_hen = []
std_hen = []
for i in length:
    df_hctsa = dict_hctsa[i]
    df_hctsa = df_hctsa.loc[model, op]
    # Mean
    avg = df_hctsa.mean()
    std = df_hctsa.std(ddof=1)
    avg_hen.append(avg)
    std_hen.append(std)
df_avg_hen = pd.DataFrame(columns=['Length', 'Mean'], data={'Length': length, 'Mean': avg_hen})
df_std_hen = pd.DataFrame(columns=['Length', 'Std'], data={'Length': length, 'Std': std_hen})

model = 'AR1_UNO' # arma

avg_ar1 = []
std_ar1 = []
for i in length:
    df_hctsa = dict_hctsa[i]
    df_hctsa = df_hctsa.loc[model, op]
    # Mean
    avg = df_hctsa.mean()
    std = df_hctsa.std(ddof=1)
    avg_ar1.append(avg)
    std_ar1.append(std)
df_avg_ar1 = pd.DataFrame(columns=['Length', 'Mean'], data={'Length': length, 'Mean': avg_ar1})
df_std_ar1 = pd.DataFrame(columns=['Length', 'Std'], data={'Length': length, 'Std': std_ar1})

model = 'HEN_SUM' # sum chaos

avg_hen_sum = []
std_hen_sum = []
for i in length:
    df_hctsa = dict_hctsa[i]
    df_hctsa = df_hctsa.loc[model, op]
    # Mean
    avg = df_hctsa.mean()
    std = df_hctsa.std(ddof=1)
    avg_hen_sum.append(avg)
    std_hen_sum.append(std)
df_avg_hen_sum = pd.DataFrame(columns=['Length', 'Mean'], data={'Length': length, 'Mean': avg_hen_sum})
df_std_hen_sum = pd.DataFrame(columns=['Length', 'Std'], data={'Length': length, 'Std': std_hen_sum})

model = 'MODA' # other deterministic

avg_moda = []
std_moda = []
for i in length:
    df_hctsa = dict_hctsa[i]
    df_hctsa = df_hctsa.loc[model, op]
    # Mean
    avg = df_hctsa.mean()
    std = df_hctsa.std(ddof=1)
    avg_moda.append(avg)
    std_moda.append(std)
df_avg_moda = pd.DataFrame(columns=['Length', 'Mean'], data={'Length': length, 'Mean': avg_moda})
df_std_moda = pd.DataFrame(columns=['Length', 'Std'], data={'Length': length, 'Std': std_moda})
```

```python
# Compute both measures as above for models = sine_stoch, ar1_uno, 
op1 = 'AC_nl_001'
op2 = 'SB_MotifTwo_diff_uu'
op3 = 'MF_steps_ahead_ar_2_6_mabserr_1'

models = ['AR1_UNO', 'HEN', 'HEN_SUM', 'MODA','SINE_STOCH']
palette_c = ["#6dbe59",  # arma
            "#b07de2",  # chaos
            "#c2ad39",  # sum chaos
            "#568eff",  # other deterministic
            "#e57548"]  # other stochastic

palette_c_mid = [
    "#549c44",  # medium green (arma)
    "#8f5eb6",  # medium purple (chaos)
    "#9c8c2b",  # medium olive (sum chaos)
    "#4170cc",  # medium blue (other deterministic)
    "#b95533"   # medium orange (other stochastic)
]


palette_c_dark = [
    "#3a6b2e",  # dark green (arma)
    "#6b3a91",  # deep purple (chaos)
    "#7a6d1c",  # muted olive (sum chaos)
    "#2c4a80",  # navy-ish blue (other deterministic)
    "#993a1f"   # brick orange (other stochastic)
]

dict_avg_op1 = {}
dict_std_op1 = {}
dict_avg_op2 = {}
dict_std_op2 = {}
dict_avg_op3 = {}
dict_std_op3 = {}
for op in [op1, op2, op3]:
    for model in models:
        avg_model = []
        std_model = []
        for i in length:
            df_hctsa = dict_hctsa[i]
            df_hctsa = df_hctsa.loc[model, op]
            # Mean
            avg = df_hctsa.mean()
            std = df_hctsa.std(ddof=1)
            avg_model.append(avg)
            std_model.append(std)
        df_avg_model = pd.DataFrame(columns=['Length', 'Mean'], data={'Length': length, 'Mean': avg_model})
        df_std_model = pd.DataFrame(columns=['Length', 'Std'], data={'Length': length, 'Std': std_model})
        if op == op1:
            dict_avg_op1[model] = df_avg_model
            dict_std_op1[model] = df_std_model
        elif op == op2:
            dict_avg_op2[model] = df_avg_model
            dict_std_op2[model] = df_std_model
        else:
            dict_avg_op3[model] = df_avg_model
            dict_std_op3[model] = df_std_model
```

```python
# Plot in subfigs one per model (dict)
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(14, 8))
axs = axs.flatten()
lst_titles = ['AR1_UNO', 'HEN', 'HEN_SUM', 'MOD1', 'SINE_MAP']
for i, model in enumerate(models):
  # plot and fill between
    axs[i].plot(dict_avg_op1[model]['Length'], dict_avg_op1[model]['Mean'], marker='o', label=model, color=palette_c[i])
    axs[i].fill_between(dict_avg_op1[model]['Length'], dict_avg_op1[model]['Mean'] - dict_std_op1[model]['Std'], dict_avg_op1[model]['Mean'] + dict_std_op1[model]['Std'], color= palette_c[i],alpha=0.2)
    axs[i].plot(dict_avg_op2[model]['Length'], dict_avg_op2[model]['Mean'], marker='o', label=model, color=palette_c_mid[i], linestyle='--')
    axs[i].fill_between(dict_avg_op2[model]['Length'], dict_avg_op2[model]['Mean'] - dict_std_op2[model]['Std'], dict_avg_op2[model]['Mean'] + dict_std_op2[model]['Std'], color= palette_c_mid[i],alpha=0.2)
    axs[i].set_title(f'{model}')
    axs[i].plot(dict_avg_op3[model]['Length'], dict_avg_op3[model]['Mean'], marker='o', label=model, color=palette_c_dark[i], linestyle=':')
    axs[i].fill_between(dict_avg_op3[model]['Length'], dict_avg_op3[model]['Mean'] - dict_std_op3[model]['Std'], dict_avg_op3[model]['Mean'] + dict_std_op3[model]['Std'], color= palette_c_dark[i],alpha=0.2)
    axs[i].set_title(lst_titles[i])
    axs[i].set_xscale('log')
    axs[i].set_ylabel('')

    if i == len(models) - 1 or i == len(models) - 2:
        axs[i].set_xlabel('Time series length')
    else:
        axs[i].set_xticklabels([])   # remove tick labels
axs[2].set_ylabel(r'Feature difference (mean $\pm$ std)')
# Reduce the space between the plots to fit better in the figure
plt.subplots_adjust(hspace=0.4)
# ADD LEGEND INSTEAD OF LAST SUBPLOT
handles = []
labels = []
line1, = axs[0].plot([], [], marker='o', color='black', label=r'$\Delta C_{1,3}(\boldsymbol{x};1)$')
line2, = axs[0].plot([], [], marker='o', color='black', linestyle='--', label=r'$p_{\text{uu}}(\boldsymbol{x})$')
line3, = axs[0].plot([], [], marker='o', color='black', linestyle=':', label=r'$\Delta \text{MAE AR(2) predictor}$')
handles.extend([line1, line2, line3])
labels.extend([line.get_label() for line in handles])
fig.legend(handles, labels,  ncol=1,  bbox_to_anchor=(1, 0))

# Save
plt.savefig(path_base+'/Robustness.svg', format='svg', bbox_inches='tight')
plt.show()
```
